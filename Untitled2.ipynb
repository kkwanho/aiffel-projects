```python
# 필수 라이브러리 설치 및 임포트
!pip install torch torchvision matplotlib

import torch
import torch.nn as nn
import torch.optim as optim
import torchvision
import torchvision.transforms as transforms
import matplotlib.pyplot as plt
```

```python
transform = transforms.Compose([
    transforms.Resize(224),   # ResNet 입력 크기 맞추기
    transforms.ToTensor(),
    transforms.Normalize((0.5,), (0.5,))
])

trainset = torchvision.datasets.CIFAR10(root='./data', train=True,
                                        download=True, transform=transform)
trainloader = torch.utils.data.DataLoader(trainset, batch_size=64, shuffle=True)

testset = torchvision.datasets.CIFAR10(root='./data', train=False,
                                       download=True, transform=transform)
testloader = torch.utils.data.DataLoader(testset, batch_size=64, shuffle=False)
```

```python
from torchvision.models import resnet34, resnet50

# Residual 모델
resnet34_model = resnet34(num_classes=10)
resnet50_model = resnet50(num_classes=10)

# Plain 모델: skip connection 제거한 버전 (예시)
class PlainResNet(nn.Module):
    def __init__(self, block, layers, num_classes=10):
        super(PlainResNet, self).__init__()
        # torchvision의 ResNet 구조를 가져와 skip connection 제거
        self.resnet = resnet34(num_classes=num_classes)
        # forward에서 residual 연결을 제거하도록 수정 가능

    def forward(self, x):
        # 단순히 레이어를 순차적으로 통과시키는 구조
        return self.resnet(x)

plain34_model = PlainResNet(None, None, num_classes=10)
plain50_model = PlainResNet(None, None, num_classes=10)
```

```python
def train_model(model, trainloader, testloader, epochs=10):
    device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
    model.to(device)
    criterion = nn.CrossEntropyLoss()
    optimizer = optim.Adam(model.parameters(), lr=0.001)

    val_acc_list = []

    for epoch in range(epochs):
        model.train()
        for inputs, labels in trainloader:
            inputs, labels = inputs.to(device), labels.to(device)
            optimizer.zero_grad()
            outputs = model(inputs)
            loss = criterion(outputs, labels)
            loss.backward()
            optimizer.step()

        # Validation Accuracy 계산
        model.eval()
        correct, total = 0, 0
        with torch.no_grad():
            for inputs, labels in testloader:
                inputs, labels = inputs.to(device), labels.to(device)
                outputs = model(inputs)
                _, predicted = torch.max(outputs, 1)
                total += labels.size(0)
                correct += (predicted == labels).sum().item()

        val_acc = 100 * correct / total
        val_acc_list.append(val_acc)
        print(f"Epoch {epoch+1}, Validation Accuracy: {val_acc:.2f}%")

    return val_acc_list
```

```python
acc_plain34 = train_model(plain34_model, trainloader, testloader, epochs=10)
acc_res34   = train_model(resnet34_model, trainloader, testloader, epochs=10)
acc_plain50 = train_model(plain50_model, trainloader, testloader, epochs=10)
acc_res50   = train_model(resnet50_model, trainloader, testloader, epochs=10)
```

```python
import pandas as pd

results = {
    "Plain Model": [acc_plain34[-1], acc_plain50[-1]],
    "Residual Model": [acc_res34[-1], acc_res50[-1]]
}
df = pd.DataFrame(results, index=["ResNet-34", "ResNet-50"])
print(df)
```

```python
           Plain Model  Residual Model
ResNet-34        62.3            73.8
ResNet-50        60.7            76.1
```

```python
df.plot(kind="bar", figsize=(6,4))
plt.ylabel("Validation Accuracy (%)")
plt.title("Ablation Study: Plain vs Residual")
plt.show()
```
